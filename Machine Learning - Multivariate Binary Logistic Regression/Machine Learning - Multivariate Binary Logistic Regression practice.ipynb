```python
##MACHINE LEARNING - MULTIVARIATE BINARY LOGISTIC REGRESSION Practice

#Imports
import matplotlib.pyplot as plt
import math

#Variables
a = []
n = 0
nof = 0
lr = []
tempcost = 0
percentchange = 0
REQ_PERCENT_CHANGE = 0.0000000001
inp = []
y_index = 0
x_mul = []
y_mul = 1
x_centroid = []
y_centroid = 0
x_min = []
x_max = []
y_min = 0
y_max = 0
LR_FACTOR = []
PLT_MARGIN = 0.05

#Get input as list of list with format [(x0,x1,x2,...,y1)] from input.txt
with open('train.txt') as inpfile:
  content = inpfile.readlines()
  n = len(content)
  temp = content[0].split()
  nof = y_index = len(temp)
  entry = [1] + [float(z) for z in temp]
  for ii in range(0, nof):
    a.append(float(0.5))
    lr.append(float(1))    
    x_mul.append(1)
    x_centroid.append(0)
    x_min.append(0)
    x_max.append(0)
    LR_FACTOR.append(2)
    
  inp.append(entry)  
  x_centroid[0] = 1

  for ii in range(1,n):
    data = [1] + [float(temp) for temp in content[ii].split()]
    inp.append(data)
    for jj in range(1,nof):
        x_centroid[jj] = x_centroid[jj] + data[jj]

for jj in range(1, nof):
  x_centroid[jj] = x_centroid[jj]/n
  
#Min and Max of Data
x_min[0] = inp[0][0]
x_max[0] = inp[0][0]
for jj in range(1,nof):
  x_min[jj] = inp[0][jj]
  x_max[jj] = inp[0][jj]
for ii in range(1, n):
  for jj in range(1,nof):
    if inp[ii][jj]<x_min[jj]:
      x_min[jj] = inp[ii][jj]
    if inp[ii][jj]>x_max[jj]:
      x_max[jj] = inp[ii][jj] 

#Used to normalize data to fit centroid in between 1(inc) and 10(ex)
def multiplier(z):
  mul = 0
  if z != 0:
    if z >= 10:
      while int(z)%10 != 0:
        mul = mul + 1
        z = z / 10
    if z < 1:
      while int(z)%10 == 0:
        mul = mul - 1
        z = z * 10 
  #Format is [1-10)*(10^mul)
  return mul

#Normalize data
x_mul[0] = 0
for jj in range(1,nof):
  x_mul[jj] = multiplier(x_centroid[jj])
for ii in range(0,n):
  for jj in range(0, nof):
    inp[ii][jj] = inp[ii][jj] / 10**int(x_mul[jj])

#Modulus function
def mod(z):
  if float(z)<0:
    return -z
  return z

#Hypothesis
def hypo(data_index):
  temp = a[0]
  for ii in range(1, nof):
    temp = temp + a[ii]*inp[data_index][ii]
  temp = 1/(1 + math.exp(-temp))
  return temp

#Cost function
def cost():
  c = 0
  for ii in range(0,n):
    if inp[ii][y_index] == 1:
      c = c - math.log(hypo(ii))
    else:
      c = c - math.log(1 - hypo(ii))
  c = c/n
  return c

#For partial derivative of Cost w.r.t a[jj]
def sumtemp(jj):
  temp = 0
  for ii in range(0,n):
    temp = temp + (inp[ii][jj]*(hypo(ii) - inp[ii][y_index]))
  return temp

#Signum
def signum(z):
  if z >= 0:
    return 1
  elif z < 1:
    return -1

#Algorithm
#Hypothesis: H(x) = 1/(1+e^(-(a0 + a1*x1 + a2*x2 + ... )))
tempcost = cost()
while True:
  #Refine weights
  tempa = []
  for jj in range(0,nof):
    tempa.append(a[jj] - lr[jj]*(sumtemp(jj)/n))
    #Learning Rate Algorithm
    if signum(a[jj]) != signum(tempa[jj]):
        lr[jj] = lr[jj] / LR_FACTOR[jj]
  
  for jj in range(0,nof):
    a[jj] = tempa[jj]

  #Cost and changes
  tempcost2 = cost()
  percent_change = (tempcost2 - tempcost)/100
  if mod(percent_change) <= REQ_PERCENT_CHANGE:
    break
  tempcost = tempcost2

#Final modifications of coefficients
for jj in range(0,nof):
  a[jj] = a[jj] * 10**(-1-x_mul[jj])

#Print
#We now have final (kinda) accurate Hypothesis
print(f"Hypothesis := 1 / (1 + e^(-1 * AX))")
print(f"AX : = ")
print(f"{a[0]:0.15f}")
for jj in range(1,nof):
  print(f"{a[jj]:0.15f} * x{jj}")
print("\n\nTest:")
            
with open("test.txt") as test:
  test_contents = test.readlines()
  m = len(test_contents)
  for kk in range(0, m):
    AX = a[0]
    x_inps = [eval(z) for z in test_contents[kk].split()]
    for ll in range(0, len(x_inps)):
      if(x_inps[ll] < x_min[ll+1]):
        x_min[ll+1] = x_inps[ll]
      if(x_inps[ll] > x_max[ll+1]):
        x_max[ll+1] = x_inps[ll]
      AX = AX + a[ll+1]*x_inps[ll]
      temporary = 1/(1+math.exp(-AX)) * 100
    if(AX >= 0):
      print(f"1 with {temporary:0.2f}% probability for input = {x_inps}")
      if nof == 2:
        plt.plot(x_inps, [1], 'rx')
      elif nof == 3:
        plt.plot(x_inps[0], x_inps[1], 'rx')
    else:
      print(f"0 with {100 - temporary:0.2f}% probability for input = {x_inps}")
      if nof == 2:
        plt.plot(x_inps, [0], 'ro')
      elif nof == 3:
        plt.plot(x_inps[0], x_inps[1], 'ro')
            
if nof == 2:
  #Plot
  x_diff = x_max[1] - x_min[1]
  X_MIN = x_min[1] - (x_diff)*PLT_MARGIN
  X_MAX = x_max[1] + (x_diff)*PLT_MARGIN
  Y_MIN = 0 - PLT_MARGIN
  Y_MAX = 1 + PLT_MARGIN
  plt.xlim(X_MIN,X_MAX)
  plt.ylim(Y_MIN,Y_MAX)
  plt.plot([float(z[1]*10**x_mul[1]) for z in inp],[z[2] for z in inp],'o')
  plt.plot([-a[0]/a[1],-a[0]/a[1]],[Y_MIN,Y_MAX],'-r')
  plt.show()
elif nof == 3:
  #Plot
  x_diff = x_max[1] - x_min[1]
  y_diff = x_max[2] - x_min[2]
  X_MIN = x_min[1] - (x_diff)*PLT_MARGIN
  X_MAX = x_max[1] + (x_diff)*PLT_MARGIN
  Y_MIN = x_min[2] - (y_diff)*PLT_MARGIN
  Y_MAX = x_max[2] + (y_diff)*PLT_MARGIN
  plt.xlim(X_MIN,X_MAX)
  plt.ylim(Y_MIN,Y_MAX)
  for kk in range(0, n):
    z = inp[kk]
    if z[3] == 1:
      plt.plot(z[1], z[2], 'bx')
    else:
      plt.plot(z[1], z[2], 'bo')
  plt.plot([X_MIN, X_MAX], [(-a[0]-a[1]*X_MIN)/a[2], (-a[0]-a[1]*X_MAX)/a[2]],'-r')
  plt.show()
```
