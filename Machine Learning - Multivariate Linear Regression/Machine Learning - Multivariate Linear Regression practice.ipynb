```python
##MACHINE LEARNING - MULTIVARIATE LINEAR REGRESSION Practice

#Imports
import matplotlib.pyplot as plt

#Variables
a = []
n = 0
nof = 0
lr = []
tempcost = 0
percentchange = 0
REQ_PERCENT_CHANGE = 0.0000000000001
inp = []
y_index = 0
x_mul = []
y_mul = 1
x_centroid = []
y_centroid = 0
x_min = []
x_max = []
y_min = 0
y_max = 0
LR_FACTOR = []
PLT_MARGIN = 0.05

#Get input as list of list with format [(x0,x1,x2,...,y1)] from input.txt
with open('input.txt') as inpfile:
  content = inpfile.readlines()
  n = len(content)
  temp = content[0].split()
  nof = y_index = len(temp)
  entry = [1] + [float(z) for z in temp]
  for ii in range(0, nof):
    a.append(float(0))
    lr.append(float(1))    
    x_mul.append(1)
    x_centroid.append(0)
    x_min.append(0)
    x_max.append(0)
    LR_FACTOR.append(2)
    
  inp.append(entry)  
  x_centroid[0] = 1

  for ii in range(1,n):
    data = [1] + [float(temp) for temp in content[ii].split()]
    inp.append(data)
    for jj in range(1,nof):
        x_centroid[jj] = x_centroid[jj] + data[jj]
        y_centroid = y_centroid + data[y_index]

for jj in range(1, nof):
  x_centroid[jj] = x_centroid[jj]/n
y_centroid = y_centroid/n
  
#Min and Max of Data
x_min[0] = inp[0][0]
x_max[0] = inp[0][0]
y_min = inp[0][y_index]
y_max = inp[0][y_index]
for jj in range(1,nof):
  x_min[jj] = inp[0][jj]
  x_max[jj] = inp[0][jj]
for ii in range(1, n):
  for jj in range(1,nof):
    if inp[ii][jj]<x_min[jj]:
      x_min[jj] = inp[ii][jj]
    if inp[ii][jj]>x_max[jj]:
      x_max[jj] = inp[ii][jj]
  if inp[ii][y_index]<y_min:
    y_min = inp[ii][y_index]
  if inp[ii][y_index]>y_max:
    y_max = inp[ii][y_index] 

#Used to normalize data to fit centroid in between 1(inc) and 10(ex)
def multiplier(z):
  mul = 0
  if z != 0:
    if z >= 10:
      while int(z)%10 != 0:
        mul = mul + 1
        z = z / 10
    if z < 1:
      while int(z)%10 == 0:
        mul = mul - 1
        z = z * 10 
  #Format is [1-10)*(10^mul)
  return mul

#Normalize data
x_mul[0] = 0
for jj in range(1,nof):
  x_mul[jj] = multiplier(x_centroid[jj])
y_mul = multiplier(y_centroid)
for ii in range(0,n):
  for jj in range(0, nof):
    inp[ii][jj] = inp[ii][jj] / 10**int(x_mul[jj])
  inp[ii][y_index] = inp[ii][y_index] / 10**int(y_mul)

#Modulus function
def mod(z):
  if float(z)<0:
    return -z
  return z

#Hypothesis
def hypo(data_index):
  temp = a[0]
  for ii in range(1, nof):
    temp = temp + a[ii]*inp[data_index][ii]
  return temp

#Cost function
def cost():
  c = 0
  for ii in range(0,n):
    x = hypo(ii) - inp[ii][y_index]
    c = c + x*x
  c = c/(2*n)
  return c

#For partial derivative of Cost w.r.t a[jj]
def sumtemp(jj):
  temp = 0
  for ii in range(0,n):
    temp = temp + (inp[ii][jj]*(hypo(ii) - inp[ii][y_index]))
  return temp

#Signum
def signum(z):
  if z >= 0:
    return 1
  elif z < 1:
    return -1

#Algorithm
#Hypothesis: H(x) = a0 + a1*x1 + a2*x2 + ...
tempcost = cost()
while True:
  #Refine weights
  tempa = []
  for jj in range(0,nof):
    tempa.append(a[jj] - lr[jj]*(sumtemp(jj)/n))
    #Learning Rate Algorithm
    if signum(a[jj]) != signum(tempa[jj]):
        lr[jj] = lr[jj] / LR_FACTOR[jj]
  
  for jj in range(0,nof):
    a[jj] = tempa[jj]

  #Cost and changes
  tempcost2 = cost()
  percent_change = (tempcost2 - tempcost)/100
  if mod(percent_change) <= REQ_PERCENT_CHANGE:
    break
  tempcost = tempcost2

#Final modifications of coefficients
for jj in range(0,nof):
  a[jj] = a[jj] * 10**(y_mul - x_mul[jj])

#Print
#We now have final (kinda) accurate Hypothesis
print(f"Hypothesis :=")
print(f"{a[0]:0.15f}")
for jj in range(1,nof):
  print(f"{a[jj]:0.15f} * x{jj}")

if nof == 2:
  #Plot
  x_diff = x_max[1] - x_min[1]
  y_diff = y_max - y_min
  X_MIN = x_min[1] - (x_diff)*PLT_MARGIN
  X_MAX = x_max[1] + (x_diff)*PLT_MARGIN
  Y_MIN = y_min - (y_diff)*PLT_MARGIN
  Y_MAX = y_max + (y_diff)*PLT_MARGIN
  plt.xlim(X_MIN,X_MAX)
  plt.ylim(Y_MIN,Y_MAX)
  plt.plot([float(z[1]*10**x_mul[1]) for z in inp],[float(z[2]*10**y_mul) for z in inp],'o')
  plt.plot([X_MIN,X_MAX],[a[0]+X_MIN*a[1],a[0]+X_MAX*a[1]],'-r')
  plt.plot([x_centroid[1]*10**x_mul[1]],[y_centroid]*10**y_mul,'ro')
  plt.show()
```
